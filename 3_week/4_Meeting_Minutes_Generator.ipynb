```python
import os
import requests
from IPython.display import Markdown, display, update_display
from openai import OpenAI
from google.colab import userdata
from huggingface_hub import login
from google.colab import drive
from transformers import AutoTokenizer, AutoModelForCausalLM, TextStreamer, BitsAndBytesConfig
import torch
```

```python
# Constants

AUDIO_MODEL = "whisper-1"
LLAMA = "meta-llama/Meta-Llama-3.1-8B-Instruct"
```

```python
# New Capability - connect this colab to my google drive

drive.mount('/content/drive')
audio_filename = "/content/drive/MyDrive/denver_extract.mp3"
```

```python
# Sign in to HuggingFace Hub

hf_token = userdata.get('HF_TOKEN')
login(hf_token)
```

```python
# Sign in to OpenAI

openai_api_key = userdata.get("OPENAI_API_KEY")
openai = OpenAI(api_key = openai_api_key)
```

```python
# Use the whisper OpenAI model to convert Audio into Text

audio_file = open(audio_filename, "rb")
transcription = openai.audio.transcriptions.create(model = AUDIO_MODEL, file = audio_file, response_format = "text")
print(transcription)
```

```python
system_message = "You are an assistant that produces minutes of meetings from transcripts, with summary, key discussion points, takeaways and action item with owners, in Markdown."
user_message = f"Below is an extract transcript of a Denver council meeting. Please write minutes in Markdown, including a summary with attendees, location and date; discussion points; takeaways; and action items with owners.\n{transcription}"

messages = [
    {'role': 'system', 'content': system_message},
    {'role': 'user', 'content': user_message}
]
```

```python
quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16)
```

```python
tokenizer = AutoTokenizer.from_pretrained(LLAMA)
tokenizer.pad_token = tokenizer.eos_token
inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")
streamer = TextStreamer(tokenizer)
model = AutoModelForCausalLM.from_pretrained(LLAMA, device_map="auto", quantization_config=quant_config)
outputs = model.generate(inputs, max_new_tokens=2000, streamer=streamer)
```

```python
response = tokenizer.decode(outputs[0])
```

```python
display(Markdown(response))
```
